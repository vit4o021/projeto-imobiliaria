# Manipulando os dados

## Criando colunas numéricas

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
dados.head()
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio'] #Criando a coluna Valor por mês
dados.head()
```

```python
dados['valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU'] #Criando a coluna Valor por ano
dados.head()
```

## Criando colunas categóricas

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro']
dados.head()
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                     dados['Quartos'].astype(str) + ' quarto(s) ' + \
                    ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.' # \ - quebra uma linha | astype - muda o tipo de dado. Nesse caso foi mudado pra string, pois não se pode concatenar string com inteiro.
dados.head()
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não") #apply - utilizado para aplicar uma determinada função em um conjunto específicos de dados (Coluna Suítes)
dados.head()
```
